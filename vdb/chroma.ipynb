```python
import chromadb
chroma_client = chromadb.Client()
collection = chroma_client.create_collection("my_documents")
```

```python
collection.add(
    documents=[
        "The SAC agent uses a maximum entropy objective",
        "WMA filter reduces action volatility by 92%",
        "Reinforcement learning requires a reward function",
        "Python is a programming language"
    ],
    ids=["doc1", "doc2", "doc3", "doc4"]
)
```

```python
# 3. Query — find documents semantically similar to your question
results = collection.query(
    query_texts=["how does action smoothing work?"],
    n_results=2
)
```

```python
print(results['documents'])
# Returns doc1 and doc2 — because they're semantically related
# even though "action smoothing" isn't in those exact words
```
